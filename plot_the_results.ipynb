# Import libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors
```

# Download dataframes

```python
start_path = "D:\\dossier important 2020\\swa_gaussian-master\\optimizer_results\\"

file_names_dic_AG_NEWS = {
    "SWAG_AG_NEWS": start_path+"SWAG_AG_NEWS.csv",
    "SWA": start_path+"SWA.csv",
    "SGD_scheduler_COSINE": start_path+"SGD_scheduler_COSINE.csv",
    "Adagrad": start_path+"Adagrad.csv",
    "SGD": start_path+"SGD.csv"
}

file_names_dic_YelpReviewFull = {
    "SWAG_YelpReviewFull": start_path+"SWAG.csv",
    "SWA_YelpReviewFull": start_path+"SWA_YelpReviewFull.csv",
    "SGD_scheduler_COSINE_YelpReviewFull": start_path+"SGD_scheduler_COSINE_YelpReviewFull.csv",
    "Adagrad_YelpReviewFull": start_path+"Adagrad_YelpReviewFull.csv",
    "SGD_YelpReviewFull": start_path+"SGD_YelpReviewFull.csv"
}
```

# Study on the AG News Dataset

```python
dic_dataframe_AG_NEWS = dict()

for method,path in file_names_dic_AG_NEWS.items():
    dic_dataframe_AG_NEWS[method] = pd.read_csv(path) 
    print(method)
    #print(dic_dataframe_AG_NEWS[method])
```

```python
i = 0
fig, ax = plt.subplots()
for method, dataframe in dic_dataframe_AG_NEWS.items():
    
    indice_color_float = np.random.random_sample(3)
    indice_color = matplotlib.colors.to_rgb(indice_color_float)
    if method == "SWAG_AG_NEWS":
        ax.plot(dataframe['train_swag_accuracy']/100, color=indice_color,label=method)
    else:
        ax.plot(dataframe['train_acc_tab'], color=indice_color,label=method)
legend = ax.legend()
plt.show()
```

```python
i = 0
fig, ax = plt.subplots()
for method, dataframe in dic_dataframe_AG_NEWS.items():
    
    indice_color_float = np.random.random_sample(3)
    indice_color = matplotlib.colors.to_rgb(indice_color_float)
    if method == "SWAG_AG_NEWS":
        ax.plot(dataframe['test_swag_accuracy']/100, color=indice_color,label=method)
    else:
        ax.plot(dataframe['valid_acc_tab'], color=indice_color,label=method)
legend = ax.legend()
plt.show()
```
